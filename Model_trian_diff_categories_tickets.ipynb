```python
from transformers import MarianMTModel, MarianTokenizer
from tqdm import tqdm
import torch

# Load model

model_name = "Helsinki-NLP/opus-mt-de-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model = model.to(device)
model.eval()

if device == "cuda":
    model.half()

# Translation function
def translate(texts, batch_size=64, max_length=128):
    translated_texts = []

    for i in tqdm(range(0, len(texts), batch_size), desc="Translating"):
        batch = texts[i:i + batch_size]
        batch = [str(x) for x in batch]  # safety

        inputs = tokenizer(
            batch,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=max_length
        ).to(device)

        with torch.no_grad():
            outputs = model.generate(**inputs)

        decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
        translated_texts.extend(decoded)

    return translated_texts

#  Extract German rows WITH index
german_series = (
    df_train_updated.loc[df_train_updated["language"] == "de", "body"]
    .dropna()
    .astype(str)
    .str.strip()
)

german_series = german_series[german_series != ""]

german_text = german_series.tolist()
german_index = german_series.index  

#  Translate

translated_german = translate(german_text)

# Assign back to DataFrame
df_train_updated.loc[german_index, "translated"] = translated_german

# Optional: keep English as-is
df_train_updated.loc[
    df_train_updated["language"] == "en", "translated"
] = df_train_updated.loc[
    df_train_updated["language"] == "en", "body"  
]


# Check

print(df_train_updated[["language", "body", "translated"]].head())
```
